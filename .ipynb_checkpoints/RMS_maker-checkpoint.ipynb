```python
import numpy as np
import numpy.ma as ma

import matplotlib.pyplot as plt
import matplotlib.colors as colors

from scipy import stats

from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits
from astropy.stats import sigma_clipped_stats
from astropy.wcs import WCS
from astropy.convolution import convolve, convolve_fft
from astropy.convolution import Gaussian2DKernel

from tqdm import tqdm
```

```python
path = r'/data2/goesaert/'
fname = 'Abell_85_Linpol_Freqmean'
hdu = fits.open(get_pkg_data_filename(path+fname+'.fits'))[0]
header = hdu.header
data = hdu.data
```

```python
mask_sources = np.load('testmask.npy')
data_masked = ma.masked_array(data, mask = mask_sources)
```

```python
plt.figure(figsize=(12,8))
plt.title('Used Map')
plt.imshow(data_masked.clip(min=0.0000001), cmap='Greys', origin='lower', norm=colors.LogNorm(), interpolation='nearest')
plt.colorbar()
plt.xlim(1200, 1400)
plt.ylim(1600, 1800)
plt.show()
```

```python
def create_circular_mask(h, w, center=None, radius=None):
    if center is None: # use the middle of the image
        center = (int(w/2), int(h/2))
    if radius is None: # use the smallest distance between the center and image walls
        radius = min(center[0], center[1], w-center[0], h-center[1])
    Y, X = np.ogrid[:h, :w]
    dist_from_center = np.sqrt((X - center[0])**2 + (Y-center[1])**2)
    mask_loc = dist_from_center >= radius
    return mask_loc

def RMS_calc(data, radius, stepsize):
    height, width = np.shape(data_masked)
    h_steps = np.arange(0, width+stepsize, stepsize)
    v_steps = np.arange(0, height+stepsize, stepsize)
    
    Mean_grid = np.zeros((len(v_steps), len(h_steps)))
    RMS_grid = np.zeros((len(v_steps), len(h_steps)))
    print(np.shape(Mean_grid))
    for n, h_step in enumerate(tqdm(v_steps)):
        for i, v_step in enumerate(h_steps):
            mask_circ = create_circular_mask(height, width, center=(h_step, v_step))
            RMS_region = ma.masked_array(data_masked, mask = mask_circ)
            Mean_grid = np.mean(RMS_region)
            RMS_grid = np.std(RMS_region)
    return Mean_grid, RMS_grid
```

```python
data_masked_nan = np.where(data_masked.mask,np.nan,data_masked)
kernel = Gaussian2DKernel(x_stddev=10, x_size=51, y_size=51)
mean = convolve(data_masked_nan, kernel, boundary='extend')
mean2 = convolve_fft(data_masked_nan, kernel)
rms = np.sqrt(convolve(data_masked_nan**2, kernel, boundary='extend'))
rms2 = np.sqrt(convolve_fft(data_masked_nan**2, kernel))
```

```python
plt.figure(figsize=(12,8))
plt.title('Mean Map')
plt.imshow(rms, vmin=0.5e-5, vmax=5e-5, cmap='Greys', origin='lower', interpolation='nearest')
plt.colorbar()
#plt.xlim(1200, 1400)
#plt.ylim(1600, 1800)
plt.show()

plt.figure(figsize=(12,8))
plt.title('Used Map')
plt.imshow(data, norm=colors.LogNorm(), cmap='Greys', origin='lower', interpolation='nearest')
plt.colorbar()
plt.xlim(1200, 1400)
plt.ylim(1600, 1800)
plt.show()

plt.figure(figsize=(12,8))
plt.title('Used Map')
plt.imshow(data/rms, cmap='Greys', origin='lower', norm=colors.LogNorm(), interpolation='nearest')
plt.colorbar()
plt.xlim(1200, 1400)
plt.ylim(1600, 1800)
plt.show()
```
