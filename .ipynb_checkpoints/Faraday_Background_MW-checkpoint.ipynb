```python
import numpy as np
import matplotlib.pyplot as plt
import h5py

from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.coordinates import ICRS, Galactic, FK5
from astropy_healpix import HEALPix
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits
from astropy.wcs import WCS

import matplotlib.colors as colors
```

```python
directory = r'/net/vdesk/data2/GoesaertW/Meerkat_Data/Abell_85'

I_hdu = fits.open(get_pkg_data_filename(directory + r'/Abell_85_aFix_pol_I_Farcsec_fcube_cor.fits'))[0]
I_cube = I_hdu.data[0]
I_wcs = WCS(I_hdu.header)
I_freqmean = np.nanmean(I_cube, axis=0)
```

```python
# open the file as 'f'
f =  h5py.File('/net/vdesk/data2/GoesaertW/faraday2020v2.hdf5', 'r')

    
faraday_sky_mean = f['faraday_sky_mean']
faraday_sky_std = f['faraday_sky_std']

size = np.shape(faraday_sky_mean)[0]
indices_healpix = np.linspace(0,size-1,size, dtype=int)

print(indices_healpix)
```

```python
hp = HEALPix(nside=512, order='ring')
gal_lon, gal_lat = hp.healpix_to_lonlat(indices_healpix)

```

```python
gal_c = SkyCoord(l=gal_lon, b=gal_lat, frame='galactic')
sky_c = gal_c.transform_to('fk5')

bounds_image = SkyCoord(["0:44:14.87 -9:55:02.78", "0:39:20.99 -9:55:02.78", "0:44:14.87 -8:43:02.57"\
                         , "0:39:20.99 -8:43:02.57"], frame=FK5, unit=(u.hourangle, u.deg), obstime="J2000")
edges_image_deg = np.array([bounds_image.ra.deg[0], bounds_image.ra.deg[3]\
                            , bounds_image.dec.deg[0], bounds_image.dec.deg[3]])
print(edges_image_deg[0])
```

```python
norm = plt.Normalize(-200, 200)

plt.figure(figsize=(25,15))
scat = plt.scatter(sky_c.ra, sky_c.dec, c=faraday_sky_mean, s=1, norm=norm)
plt.scatter(bounds_image.ra, bounds_image.dec, color='red')
plt.colorbar(scat)
plt.xlabel('ra [deg]')
plt.ylabel('dec [deg]')
plt.xlim(0,360)
plt.ylim(-90,90)
plt.show()
```

```python
mask_image = np.array((sky_c.ra<bounds_image.ra[0])&(sky_c.ra>bounds_image.ra[3])\
                         &(sky_c.dec>bounds_image.dec[0])&(sky_c.dec<bounds_image.dec[3]),dtype=bool)
print(len(np.argwhere(mask_image==True)))

faraday_image_mean = faraday_sky_mean[mask_image]
faraday_image_std = faraday_sky_std[mask_image]
sky_c_image = sky_c[mask_image]
```

```python
norm2 = plt.Normalize(-15, 1)
norm3 = plt.Normalize(4, 10)

fig, ax = plt.subplots(1,2,figsize=(20, 8))

ax[0] = plt.subplot(projection=I_wcs, slices=('x', 'y', 0, 0))
scat1 = ax[0].scatter(sky_c_image.ra, sky_c_image.dec, c=faraday_image_mean, norm=norm2, s=100, transform=ax[0].get_transform('world'))
ax[0].imshow(I_freqmean.clip(min=0.000001), cmap='gray', norm=colors.LogNorm())
fig.colorbar(scat1, ax=ax[0], label='Galactic RM [rad m$^{-2}$]')
ax[0].set_xlabel('ra [deg]')
ax[0].set_ylabel('dec [deg]')
'''
ax[1] = plt.subplot(projection=I_wcs, slices=('x', 'y', 0, 0))
scat2 = ax[1].scatter(sky_c_image.ra, sky_c_image.dec, c=faraday_image_std, norm=norm3, s=100, transform=ax[1].get_transform('world'))
ax[1].imshow(I_freqmean.clip(min=0.000001), cmap='gray', norm=colors.LogNorm())
fig.colorbar(scat2, ax=ax[1], label='Galactic $\sigma_{\mathrm{RM}}$ [rad m$^{-2}$]')
ax[1].set_xlabel('ra [deg]')
ax[1].set_ylabel('dec [deg]')'''
```

```python
RM_mean = np.mean(faraday_image_mean)
RM_std = np.mean(faraday_image_std)

print(RM_mean, RM_std)
```
